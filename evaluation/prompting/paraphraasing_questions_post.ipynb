```python
import random
import concurrent.futures  # for multithreading
import multiprocessing  # for multiprocessing
import matplotlib.pyplot as plt
import numpy as np

from together_ai_common import *
```

```python
massive_dump_dir = f'{PROJECT_PATH}/data/paraphrased'
```

```python
data = []
for f in os.listdir(massive_dump_dir):
    if f.endswith('.json'):
        with open(f'{massive_dump_dir}/{f}', 'r') as f:
            data_d = json.load(f)
            data.append(data_d)
# save_jsonl(data, f'{PROJECT_PATH}/data/test_data.paraphrased.jsonl')
```

```python
data_by_id = {d[OUT_OBJ_ID]: d for d in data}
```

```python
note_tokens = ['\nNote', "(Note"]
note_statements = []
questionable = []
for qid, d in data_by_id.items():
    for key in [OUT_OBJ_QUESTION_PARAPHRASED, OUT_OBJ_INITIAL_STATE_NL_PARAPHRASED]:
        for note_token in note_tokens:
            if note_token in d[key]:
                start_id = d[key].find(note_token)
                
                note_statements.append(d[key][start_id:])
                print('-before')
                print(d[key])
                d[key] = d[key][:start_id].strip()
                print('-after')
                print(d[key])
                print()
            print('-------')
```

```python
stuff_to_remove = [' I ']
questionable = []
for qid, d in data_by_id.items():
    for stuff in stuff_to_remove:
        if stuff in d[OUT_OBJ_QUESTION_PARAPHRASED] or stuff in d[OUT_OBJ_INITIAL_STATE_NL_PARAPHRASED]:
            questionable.append(d[OUT_OBJ_ID])
            break
```
