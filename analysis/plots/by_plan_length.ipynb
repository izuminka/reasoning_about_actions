```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
from collections import defaultdict

# Vintage-inspired colors
vintage_colors = [
    '#d45d00',  # Burnt orange
    '#729fcf',  # Lighter blue (State Tracking)
    '#4f7942',  # Darker earthy green (Action Executability)
    '#2f3e46',  # Dark blue-grey
    '#e0c079',  # Dull yellow
    '#9a031e',  # Deep red
    '#3d405b',  # Soft navy blue
    '#e36414',  # Bright orange
    '#76c893',  # Vibrant mint green
]



TO_PRETTY = {
    'object_tracking': 'Object Trk.',
    'fluent_tracking': 'Fluent Trk.',
    'state_tracking': 'State Trk.',
    'action_executability': 'Action Exec.',
    'effects': 'Effects of Ac.',
    'numerical_reasoning': 'Num. Reas.',
    'hallucination': 'Halluc.',
    'composite': 'Composite'
}

FONT_SIZE = 17

model_names = ['gpt-4o', 'llama_70b', 'llama_8b']
plan_lengths = [1, 10, 19]  # Assuming these are the plan lengths
QUESTION_CATEGORIES = ['fluent_tracking', 'state_tracking', 'action_executability', 'effects', 'numerical_reasoning', 'composite']

# Set up figure and axes for 3 side-by-side plots
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(13, 4.5))

for i, model_name in enumerate(model_names):
    all_data = open_jsonl(f'{CODE_PATH}/analysis/zero_shot/{model_name}.all.jsonl')
    all_data = [d for d in all_data if d[OUT_OBJ_DOMAIN_NAME] not in ['goldminer', 'zenotravel', 'miconic', 'npuzzle', 'logistics']]

    results_by_plan_length = {}
    error_by_plan_length = {}
    
    for plan_length in plan_lengths:
        results_by_cat = defaultdict(list)
        error_by_cat = defaultdict(list)
        for question_category in QUESTION_CATEGORIES:
            scores = [d[IS_RESPONSE_CORRECT_KEY] == TRUE_ANSWER for d in all_data if d['question_category'] == question_category and d['plan_length'] == plan_length and d[IS_RESPONSE_CORRECT_KEY] in (TRUE_ANSWER, FALSE_ANSWER)]
            results_by_cat[question_category] = np.mean(scores)
            error_by_cat[question_category] = scipy.stats.sem(scores)
        results_by_plan_length[plan_length] = results_by_cat
        error_by_plan_length[plan_length] = error_by_cat

    # Plotting for each question category with vintage colors
    for idx, category in enumerate(QUESTION_CATEGORIES):
        y = [results_by_plan_length[plan_length][category] for plan_length in plan_lengths]
        err = [error_by_plan_length[plan_length][category] for plan_length in plan_lengths]
        ax[i].errorbar(plan_lengths, y, yerr=err, capsize=3, fmt="o-", label=TO_PRETTY[category], linewidth=2, color=vintage_colors[idx % len(vintage_colors)])

    # Set labels and titles for the subplots
    ax[i].set_ylim([0.15, 1])
    ax[i].set_xticks(plan_lengths)
    ax[i].set_xlabel('Action-Sequence Length', fontsize=FONT_SIZE)
    ax[i].tick_params(axis='both', which='major', labelsize=FONT_SIZE)

    # Remove y-ticks and y-label for the second and third subplots
    if i > 0:
        ax[i].set_yticks([])  # Remove y-ticks
        ax[i].set_ylabel('')   # Remove y-label

    # Add labels below each plot
    model_title = {'gpt-4o': '(a) GPT-4o', 'llama_70b': '(b) Llama-3.1-70B-Inst', 'llama_8b': '(c) Llama-3.1-8B-Instruct'}
    ax[i].text(0.5, -0.36, model_title[model_name], ha='center', va='center', transform=ax[i].transAxes, fontsize=FONT_SIZE)

# Set y-label only for the first subplot
ax[0].set_ylabel('Accuracy', fontsize=FONT_SIZE)

# Add a single legend for the entire figure, positioned above and centered with reduced space between labels
handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', fontsize=FONT_SIZE-1, ncol=6, bbox_to_anchor=(0.529, 1.02), frameon=False,
           handletextpad=0.5, columnspacing=0.7)  # Adjust these values to reduce spacing between labels

# Adjust layout and save the figure
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('by_plan_len.vintage_style.pdf', transparent=True, format="pdf", bbox_inches='tight')

# Show the plot
plt.show()
```
